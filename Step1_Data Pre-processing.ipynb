```python
import pandas as pd
```

```python
dataset=pd.read_csv("Churn Prediction.csv")
```

```python
ds=dataset.drop(["is_trial.1", 'churn_flag.1','account_id', 'subscription_id', 'plan_tier.1',"end_date",'signup_date','churn_date',"churn_event_id"], axis=1)
```

```python
ds.head(3)
```

```python
ds.columns
```

```python
ds.rename(columns={'start_date':"Account creation date", "seats":"Licensed user count", 
                   'mrr_amount': "Monthly revenue",'arr_amount':"Annual revenue", 'is_trial': "Trial status",
                 'preceding_upgrade_flag':"Had upgrade within 90 days", 'preceding_downgrade_flag': "Had downgrade within 90 days" }, inplace=True)
```

```python
ds.columns
```

```python
ds.head(3)
```

```python
ds.isna().sum()
```

```python
ds.info()
```

```python
ds["refund_amount_usd"] = ds["refund_amount_usd"].fillna(0)
```

```python
ds["reason_code"]= ds["reason_code"].fillna("Not Available")
```

```python
ds["feedback_text"]=ds["feedback_text"].fillna("Not Available")
```

```python
ds.isnull().sum()
```

```python
nan_rows = ds[ds['is_reactivation'].isna()]
nan_rows
```

```python
ds["Had upgrade within 90 days"]=ds["Had upgrade within 90 days"].fillna("Unknown")
ds["Had downgrade within 90 days"]=ds["Had downgrade within 90 days"].fillna("Unknown")
```
